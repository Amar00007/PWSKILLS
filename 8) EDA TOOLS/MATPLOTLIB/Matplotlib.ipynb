```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,200)
```

```python
x
```

```python
y=np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
y=np.cos(x)
```

```python
plt.plot(x,y)
```

```python
plt.xlabel('this is my x data')
plt.ylabel('this is my cos data')
plt.title('cos graph')
plt.plot(x,y,'--r')
```

```python
plt.figure(figsize=(6,2))
plt.plot(x,y,'--r')
plt.xlabel('this is my x data')
plt.ylabel('this is my cos data')
plt.title('cos graph')
plt.grid()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y)
```

```python
colours=np.random.rand(50)
sizes=1000*np.random.rand(50)
plt.scatter(x,y,c=colours,s=sizes,alpha=.8)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('random data')
```

```python
x=['a','b','c','d','e']
```

```python
y=np.random.rand(5)
plt.bar(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('bar plot')
```

```python
y=np.random.rand(5)
plt.barh(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('bar plot')
```

```python
colours=np.random.rand(50)
sizes=1000*np.random.rand(50)
plt.scatter(x,y,c='b',s=sizes,alpha=.8,marker='x')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('random data')
```
